```python
print("HOW MUCH IS EACH PLATFORM USED?\n")

userNumber = []

for platform in df['Platform'].unique():
    print(platform, ">>", len(df[df['Platform'] == platform]))
    userNumber.append(len(df[df['Platform'] == platform]))
```

```python
print("HOW MUCH IS EACH EMOJI USED?\n")

contextNumber = []
for context in df['Context'].unique():
    print(context, ">>", len(df[df['Context'] == context]))
```

```python
colors = ['yellow','purple','darkblue','blue','black','green']

plt.bar(df['Platform'].unique(), userNumber, color = colors)
plt.title("PLATFORM USER COUNT GRAPH\n")
plt.xlabel("\nPLATFORMS")
plt.ylabel("USER NUMBER\n")

plt.show()
```

```python
minUser = np.min(df['User Age'])
maxUser = np.max(df['User Age'])
print("PLATFORM USER AGE RANGE: [",minUser, "-",maxUser,"]")
```

```python
platformUserAgeMeans = []

for platform in df['Platform'].unique():
    platformUserAgeMeans.append(round(df[df['Platform'] == platform]['User Age'].mean(), 1)) 
    
platformNames = ['Snapchat','Instagram','Facebook','Twitter','TikTok','WhatsApp']

for i in range(len(platformNames)):
    platformNames[i] += ("\n" + str(round(platformUserAgeMeans[i], 1)))

plt.bar(platformNames, platformUserAgeMeans, color = 'darkgray')  
plt.title("\nUSER AGE AVERAGE OF PLATFORMS\n")
plt.xlabel("\nPLATFORMS")
plt.ylabel("USER RATIO\n")

plt.show()
```

```python
maleUserNumber, femaleUserNumber = [], []

for platform in df['Platform'].unique():
    maleUserNumber.append(len((df[(df['Platform'] == platform) & (df['User Gender'] == 'Male')])))
    femaleUserNumber.append(len((df[(df['Platform'] == platform) & (df['User Gender'] == 'Female')])))    
```

```python
bar_width = 0.35
index = np.arange(len(df['Platform'].unique()))

fig, ax = plt.subplots(figsize = (10, 6))

bar1 = ax.bar(index - bar_width / 2, maleUserNumber, bar_width, label = 'Male', color = 'blue')
bar2 = ax.bar(index + bar_width / 2, femaleUserNumber, bar_width, label = 'Female', color = 'hotpink')

ax.set_title("PLATFORM USER GRAPH\n")
ax.set_xlabel("\nPLATFORMS")
ax.set_ylabel("USER NUMBER\n")
ax.set_xticks(index)
ax.set_xticklabels(df['Platform'].unique())
ax.legend()

plt.show()
```
